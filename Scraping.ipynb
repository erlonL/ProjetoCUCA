```python
import csv
dicionario_salas = {}
dicionario_tipo = {}
dicionario_local = {}
with open('Salas.csv', newline='') as arquivo_csv:
    leitor_csv = csv.DictReader(arquivo_csv)
    for linha in leitor_csv:
        bloco = linha['Bloco']
        sala = linha['Sala']
        capacidade = linha['Capacidade']
        tipo = linha['Tipo']
        chave = f"{bloco} {sala} "
        dicionario_salas[chave] = int(capacidade)
        dicionario_tipo[chave] = tipo
        dicionario_local[chave] = bloco
print(dicionario_salas)
print(dicionario_tipo)
```
